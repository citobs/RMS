###  확률적 경사하강법(Gradien Descent)

```python
from sklearn.linear_model import SGDRegressor #SGD:확률적 경사하강법
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('../dataset/LinearRegressionData.csv')
dataset.head()

# 독립변수와 종속변수를 분리하자!
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values


# data분리
from sklearn.model_selection import train_test_split
#튜플형태로 값이 4개로 반환해준다.

#훈련세트를 80으로 테스트 세트를 20으로 분류
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)



```

```python
#sgr모델 생성(박복횟수 epoch 정해주기:max_iter, eta0:학습률, )
sr = SGDRegressor(max_iter=1000, eta0=0.001, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train, color='blue') #산점도 그래프
plt.plot(X_train,sr.predict(X_train),color='green') #선그래프
plt.title('Score By hours(train data,SGD)') #제목
plt.xlabel('hours') #x축이름
plt.ylabel('score') #y축 이름
plt.show()
```

```python
sr.coef_, sr.intercept_
```

```python
# 모델 평가
sr.score(X_test, y_test)
```
